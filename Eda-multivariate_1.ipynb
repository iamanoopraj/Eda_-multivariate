```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('LoanApproval-EDA.csv')
data.head()
```

```python
df = data.copy()
```

```python
df.drop(['Unnamed: 0','loan_id'],axis=1,inplace=True)
```

```python
df.head()
```

```python
categorical = []
numerical = []

for col in df.columns:
    if df[col].dtype=='O':
        categorical.append(col)
    else:
        numerical.append(col)
```

```python
categorical
```

```python
numerical
```

# Universal Analysis:
* Study one column at a time.

```python
# catcategorical: count of values, aggregation of values
categorical
```

```python
labels = df['education'].value_counts().index
values = df['education'].value_counts().values
```

```python
sns.barplot(x=labels,y=values)

for i in range(len(labels)):
    plt.annotate(text=values[i],xy=(i,values[i]))

plt.yticks(np.arange(0,3000,500))
plt.show()
```

```python
for col in categorical:
    plt.figure()
    plt.title(col)
    labels = df['col'].value_counts().index
    values = df['col'].value_counts().values

    sns.barplot(x=labels,y=values)
    for i in range(len(labels)):
        plt.annotate(text=values[i],xy=(i,values[i]))
    plt.yticks(np.arange(0,3000,500))
    plt.show()
    
```

```python
numbers = [1,2,3,4,5,6]
# "A1,A1..."
def convert(num):
    return 'A'+str(num)
```

```python
convert(1)
```

```python
list(map(convert,numbers))
```
